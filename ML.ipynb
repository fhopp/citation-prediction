```python
from sklearn import svm
from sklearn import datasets
from sklearn.naive_bayes import MultinomialNB
from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn import linear_model
from sklearn import tree

def train(data, targets):
    """
    data: list of list describing the sample
    targets: list of label
    
    e.g. 
    data = [[1,2,3], [3,4,5]]
    targets = [0,1]
    """
    
    clf = OneVsOneClassifier(LinearSVC(random_state=0))
    #clf = linear_model.SGDRegressor()
    #clf = svm.SVC()
    #clf = MultinomialNB()
    clf.fit(data,targets)
    
    return clf

def test(data, targets, model):
    """
    classifies the data and return accuracy
    """
    predict = model.predict(data)
    correct = [1 for i in range(len(predict)) if predict[i] == targets[i] ]
    return sum(correct)/float(len(targets))
```

```python
path_train_data = "data/training_data.txt"

f = open(path_train_data, 'r', encoding="Latin-1")
train_data = []
train_label = []
inner_list = []
for line in f:
    line_content = line.split("\n")
    left_right = line_content[0].split(" = ")
    if len(left_right) == 2:
        left = left_right[0]
        right = left_right[1]
        if left == 'label':
            #print(inner_list)
            train_data.append(inner_list)
            inner_list = []
            #train_label.append(float(((right.split("{")[1]).split("}"))[0]))
            if ((right.split("{")[1]).split("}"))[0] == 'low':
                train_label.append(0)
            elif ((right.split("{")[1]).split("}"))[0] == 'medium': 
                train_label.append(1)
            else:
                train_label.append(2)
        else:
            if left == 'title':
                for v in (((right.split("{")[1]).split("}"))[0].split(",")):
                    inner_list.append(float(v))
                    #print(v)
                
            elif left != 'year' and left!= 'id':
                inner_list.append(float(((right.split("{")[1]).split("}"))[0]))
                
#print(train_data)
#print(train_label)
```

```python
path_test_data = "data/test_data.txt"

f = open(path_test_data, 'r', encoding="Latin-1")
test_data = []
test_label = []
inner_list = []
for line in f:
    line_content = line.split("\n")
    left_right = line_content[0].split(" = ")
    if len(left_right) == 2:
        left = left_right[0]
        right = left_right[1]
        if left == 'label':
            #print(inner_list)
            test_data.append(inner_list)
            inner_list = []
            #test_label.append(float(((right.split("{")[1]).split("}"))[0]))
            if ((right.split("{")[1]).split("}"))[0] == 'low':
                test_label.append(0)
            elif ((right.split("{")[1]).split("}"))[0] == 'medium': 
                test_label.append(1)
            else:
                test_label.append(2)
        else:
            if left == 'title':
                for v in (((right.split("{")[1]).split("}"))[0].split(",")):
                    inner_list.append(float(v))
                    #print(v)
                
            elif left != 'year' and left!= 'id':
                inner_list.append(float(((right.split("{")[1]).split("}"))[0]))
```

```python
import numpy as np
clf = train(np.array(train_data), np.array(train_label))
```

```python
print(test(np.array(test_data), np.array(test_label), clf))
```
